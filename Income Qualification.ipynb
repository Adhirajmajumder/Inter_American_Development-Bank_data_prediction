```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
X_features = train_dropna.drop(["Target","Id","dependency","idhogar","idhogar","edjefe","edjefa"],axis=1)
Y_target = train_dropna["Target"]
```

```python
kfold = KFold(n_splits=10,shuffle=True,random_state=0)
```

```python
clf = DecisionTreeClassifier()
scoring='accuracy'
scores = cross_val_score(clf,X_features,Y_target,cv=kfold,scoring=scoring)
print(scores)
```

```python
import numpy as np
```

```python
print(round(np.mean(scores)*100,2))
```

```python
model = clf.fit(X_features,Y_target) 
```
